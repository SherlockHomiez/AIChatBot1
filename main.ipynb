```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from transformers import AutoTokenizer
from transformers import AutoModel
import chromadb
import torch

# Initialize the Chroma client
client = chromadb.Client()

# Create a collection
collection_name = "EmbeddingCH3_Test3"
collection = client.create_collection(collection_name)

# Split the dataset into chunks
splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=10,
    is_separator_regex=False,
)
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
model = AutoModel.from_pretrained('bert-base-uncased')

def generateEmbeddings(data):
    chunkedParts = splitter.split_text(data[0])
    print(f"Number of chunks: {len(chunkedParts)}")
    
    vectors = []
    for doc in chunkedParts:
        tokenized = tokenizer(
            doc,
            padding=True,
            truncation=True,
            add_special_tokens=True,
            return_attention_mask=True, 
            return_tensors="pt"
        )

        with torch.no_grad():
            embedding = model(**tokenized)
        vector = embedding.last_hidden_state.mean(dim=1).squeeze().numpy()  # Average pooling over the tokens
        vectors.append(vector)  
    
    print(f"Number of embeddings generated: {len(vectors)}")
    return chunkedParts, vectors  # Ensure both values are returned

# Generate embeddings for dataset
dataset = ["hello. My name is Manit.hello. My name is Manit.hello. My name is Manit.hello. My name is Manit.hello. My name is Manit.hello. My name is Manit.hello. My name is Manit.hello. My name is Manit.hello. My name is Manit.hello. My name is Manit.hello. My name is Manit.hello. My name is Manit.hello. My name is Manit.hello. My name is Manit.hello. My name is Manit.hello. My name is Manit.hello. My name is Manit.hello. My name is Manit."]



# Perform a similarity search

```
